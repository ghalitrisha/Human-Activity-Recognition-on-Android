```python
import src.get_models as models
```

```python
input_shape = Xtrain.shape[1:]
num_classes = len(labels)
input_shape, num_classes
```

```python
baseline_model = models.get_baseline_model(input_shape, num_classes)
```

```python
baseline_model.summary()
```

```python
import keras
```

```python
baseline_model.compile(optimizer=keras.optimizers.Adam(learning_rate=0.003),
          loss='sparse_categorical_crossentropy',
          metrics=['accuracy'])
```

```python
 baseline_model_history = baseline_model.fit(Xtrain, ytrain, epochs=10, batch_size=512,
                          validation_split=0.2, verbose=1, callbacks=[
        keras.callbacks.EarlyStopping(monitor="val_accuracy", patience=3, restore_best_weights=True),
        keras.callbacks.ReduceLROnPlateau(monitor="val_loss", factor=0.5, patience=2)
    ])
```

```python
baseline_model_history.history['val_accuracy']
```

```python
baseline_model_loss, baseline_model_accuracy = baseline_model.evaluate(
    xtest, ytest)
```

```python
baseline_model_accuracy, baseline_model_loss
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(baseline_model_history.history['accuracy'], label='accuracy')
plt.plot(baseline_model_history.history['val_accuracy'], label = 'val_accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.ylim([0.0, 1])
plt.legend(loc='lower right')
```
