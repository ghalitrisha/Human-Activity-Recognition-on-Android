```python
autoencoder = models.get_autoencoder()
```

```python
autoencoder.compile(optimizer=keras.optimizers.Adam(learning_rate=1e-3), loss='mse')
```

```python

callbacks = [
    keras.callbacks.EarlyStopping(monitor="val_loss", patience=20, restore_best_weights=True),
    keras.callbacks.ReduceLROnPlateau(monitor="val_loss", factor=0.5, patience=8)
]

history = autoencoder.fit(
    Xtrain, Xtrain,          
    batch_size=64,
    epochs=100,
    validation_split=0.1, 
    callbacks=callbacks,
    verbose=1
)
```

```python
reconstructed_data = autoencoder.predict(xtest)
mse_normal = np.mean(np.power(xtest - reconstructed_data, 2), axis=1)

# Set threshold for anomaly detection
threshold = np.percentile(mse_normal, 98) #you can adjust this percentile value
```

```python
# Suppose you kept the last 15% as val via validation_split
# If you have explicit Xval, use that instead:
Xval = Xtrain[int(0.85*len(Xtrain)):]   # aligns with validation_split
Xval_pred = autoencoder.predict(Xval, batch_size=256, verbose=0)

# Per-window reconstruction error (MSE)
val_err = np.mean((Xval - Xval_pred)**2, axis=(1,2))

# Pick threshold as a high quantile; tune q based on desired sensitivity
q = 98   # 99.5% quantile
threshold = np.percentile(val_err, q)
print("Chosen threshold:", threshold)
```

```python
def anomaly_score(ae, X):
    Xhat = ae.predict(X, batch_size=256, verbose=0)
    return np.mean((X - Xhat)**2, axis=(1,2))  # reconstruction MSE

# Example on test windows
scores = anomaly_score(autoencoder, xtest)
flags = (scores > threshold).astype(np.int32)  # 1 = anomaly, 0 = normal

print("Anomaly rate:", flags.mean())
```

```python
combined_mse = np.concatenate([val_err, scores])
combined_labels = np.concatenate([np.zeros(len(val_err)), np.ones(len(scores))])
```

```python
from sklearn.metrics import precision_score, recall_score, f1_score

# Evaluate the model
y_pred = [1 if e > threshold else 0 for e in combined_mse]

precision = precision_score(combined_labels, y_pred)
recall = recall_score(combined_labels, y_pred)
f1 = f1_score(combined_labels, y_pred)

print(f"Precision: {precision}")
print(f"Recall: {recall}")
print(f"F1-score: {f1}")
```
